**Problem 2**. Analyze the Carbon_West data file using the following code.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
carbon_west = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Carbon_West.dat")
```

Part a. Construct a histogram of CO2 emissions. 

```python
carbon_west['CO2'].hist(edgecolor='black')
plt.title('Histogram of CO2')
plt.xlabel('ColumnName')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

Part b. Find the mean, median, and standard deviation. 

```python
print("Mean:", carbon_west['CO2'].mean())
print("Median:", carbon_west['CO2'].median())
print("Standard Deviation:", carbon_west['CO2'].std())
```

**Problem 3**. Read in the murder data set using the following code. The variable murder is the murder rate (per 100,000 population) for each state in the U.S. in 2017 according to the FBI Uniform Crime Reports. At first, do not use the observation for D.C. (DC). 

```python
murders = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Murder.dat")
```

Part a. Find the mean and standard deviation for murders without D.C.

```python
murders_no_dc = murders[murders['state'] != 'DC']

mean_murders = murders_no_dc['murder'].mean()
std_murders = murders_no_dc['murder'].std()

print("Mean:", mean_murders)
print("Standard Deviation:", std_murders)
```

Part b. Find the five-number  summary without D.C. 

```python
min_murders = murders_no_dc['murder'].min()
q1_murders = murders_no_dc['murder'].quantile(0.25)
median_murders = murders_no_dc['murder'].median()
q3_murders = murders_no_dc['murder'].quantile(0.75)
max_murders = murders_no_dc['murder'].max()

print("Minimum:", min_murders)
print("Q1:", q1_murders)
print("Median:", median_murders)
print("Q3:", q3_murders)
print("Maximum:", max_murders)
```

Part c. Now include the observation for D.C. and recalculate the mean, median, and five-number summary. What is affected more by this outlier: The mean or the median? The range or the inter-quartile range?

```python
mean_murders_with_dc = murders['murder'].mean()
std_murders_with_dc = murders['murder'].std()

min_murders_with_dc = murders['murder'].min()
q1_murders_with_dc = murders['murder'].quantile(0.25)
median_murders_with_dc = murders['murder'].median()
q3_murders_with_dc = murders['murder'].quantile(0.75)
max_murders_with_dc = murders['murder'].max()

print("Mean:", mean_murders_with_dc)
print("Standard Deviation:", std_murders_with_dc)

print("\nMinimum:", min_murders_with_dc)
print("Q1:", q1_murders_with_dc)
print("Median:", median_murders_with_dc)
print("Q3:", q3_murders_with_dc)
print("Maximum:", max_murders_with_dc)

# The mean is affected more by the outlier than the median.
# The range is affected more by the outlier than the inter-quartile range.
```

**Problem 4**. Read in the Houses data file using the following code.  The variables include price (thousands of dollars), size (square feet), tax bill (dollars), number of bathrooms, number of bedrooms, and whether the house is new (1 = yes, 0 = no) for 100 home sales in Gainesville, Florida. Let
y = selling price and x = size of home.

```python
houses = pd.read_fwf("https://stat4ds.rwth-aachen.de/data/Houses.dat")
```

Part a. Construct a scatterplot between square feet and price of a home. 

```python
plt.scatter(houses['size'], houses['price'])
plt.xlabel('Size (square feet)')
plt.ylabel('Price (thousands of dollars)')
plt.title('Scatterplot of Price vs. Size')
plt.show()
```

Part b. Find the correlation between the two variables.

```python
correlation = houses['size'].corr(houses['price'])
print("Correlation:", correlation)
```

Part c. Find and plot the regression line on top of the scatterplot. 

```python
sns.lmplot(x='size', y='price', data=houses)
plt.xlabel('Size (square feet)')
plt.ylabel('Price (thousands of dollars)')
plt.title('Scatterplot of Price vs. Size with Regression Line')
plt.show()
```

Part d. Create a side by side boxplot of price comparing homes that are new and homes that are not new. 

```python
sns.boxplot(x='new', y='price', data=houses)
plt.xlabel('New (1 = yes, 0 = no)')
plt.ylabel('Price (thousands of dollars)')
plt.title('Boxplot of Price by New')
plt.show()
```
